```python
#Import Dependencies
import pandas as pd
import matplotlib as plt
```

```python
#Defining Census Data Income 2014-2018
file = "../census_data_income/Median_Household_Income_by_State.csv"

# Read the data
income_df = pd.read_csv(file)
income_df.head()
```

```python
len(income_df["Median Household Income by State"])
```

```python
# # Grab the data contained within the "California" row and the "Unnamed: 1" column
income_df[["Median Household Income by State","Unnamed: 1"]]
```

```python
#rename columns
income_df.rename(columns={'Median Household Income by State':'State'},inplace=True)
income_df.rename(columns={'Unnamed: 1': 'Income'},inplace=True)
income_df
```

```python
# # Grab the data contained within the "California" row and the "Income" column
income_df[["State","Income"]]
```

```python
#pull out California income data 
income_df_edit=income_df[["State","Income"]][income_df[["State","Income"]]["State"]=="California"]
income_df_edit
```

```python
#Look at states with highest incomes
income_df[["State","Income"]]
income_df.sort_values(['Income'], ascending=False).groupby('Income').head(10)
income_df.drop(income_df.index[0])
income_df[["State","Income"]]
```

```python
#create bar chart showing income by state
income_df.plot(kind="bar", figsize=(10,5))
plt.title("Median Income by State 2014-2018")
plt.xlabel("State")
plt.ylabel("Income")
plt.show()
```
