```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census

# Census API Keys
from config import (api_key)
```

```python
#B01003_001E Population
#B23025_005E Unemployement count
#B02001_001E Race
# 2013 Data
year = [2013, 2014, 2015, 2016, 2017, 2018]
c = Census(api_key, year=year)

census_data = c.acs5.get(("NAME", "B01003_001E", "B23025_005E", "B02001_001E"), {'for': 'state:06, 08, 12, 16, 41, 48'})
    
#Convert to DataFrame
census_pd = pd.DataFrame(census_data)

census_pd_2013 = census_pd.rename(columns={"B01003_001E": "Population",
                                      "B23025_005E": "Unemployment count",
                                      "B02001_001E": "Race",
                                      "NAME": "Name", "state": "State"})

#Using DataFrame.insert() to add a column 
census_pd_2013.insert(0, "Year", 2013, True)
```
